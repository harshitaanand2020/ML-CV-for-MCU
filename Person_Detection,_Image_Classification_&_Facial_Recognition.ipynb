```python
# Training Data for Facial Recognition
from google.colab import drive
drive.mount('/content/drive')

PATH = "/content/drive/MyDrive/Harshita/TrainingModel/Data1"
images = glob(os.path.join(PATH, "*.jpg"))
print(images)
```

```python
# Code to resize image before they are given as input to the ML model 
import os

WIDTH = 32    # Input to the model Image Width
HEIGHT = 32   # Input to the model Image Height

def proc_images(image):
    """
    Returns two arrays: 
        x is an array of resized images
        y is an array of labels
    """
    x = [] # images as arrays
    y = [] # labels Infiltration or Not_infiltration
    for img in image:
        base = os.path.basename(img)
        y.append(int(base[0]))        

        # Read and resize image
        full_size_image = cv2.imread(img)
        try:
          x.append(cv2.resize(full_size_image, (WIDTH,HEIGHT), interpolation=cv2.INTER_AREA))
        except:
          continue
    return np.array(x)/255.0, np.array(y)
```

```python
x, y = proc_images(images)
```

```python
# CNN (given for reference)
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(2, (2,2), activation='relu', input_shape=(32, 32, 3)), # Here input is given as 32 x 32 with 3 channels
    tf.keras.layers.MaxPooling2D(1,1),
    tf.keras.layers.Conv2D(2, (2,2), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),         # You can change this number 64. Play around with it to reduce model size
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(x, y, batch_size=32, epochs= 16, verbose=1)      # You can change epochs.
```

```python
MODELS_DIR = 'models/'
if not os.path.exists(MODELS_DIR):
    os.mkdir(MODELS_DIR)
MODEL_TF = MODELS_DIR + 'model'
MODEL_NO_QUANT_TFLITE = MODELS_DIR + 'model_no_quant.tflite'
MODEL_TFLITE = MODELS_DIR + 'model.tflite'
MODEL_TFLITE_MICRO = MODELS_DIR + 'model.cc'
model.save(MODEL_TF)
converter = tf.lite.TFLiteConverter.from_saved_model(MODEL_TF)
```

```python
def representative_dataset():
    a = [] # images as arrays
    for img in images:
        base = os.path.basename(img)        
        # Read and resize image
        full_size_image = cv2.imread(img)
        img = cv2.resize(full_size_image, (WIDTH,HEIGHT), interpolation=cv2.INTER_AREA)
        img = img / 255.0
        img = img.astype(np.float32)
        a.append(img)
    a = np.array(a)
    img = tf.data.Dataset.from_tensor_slices(a).batch(1)
    for i in img.take(4):
        yield [i]

converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
converter.inference_input_type = tf.int8
converter.inference_output_type = tf.int8

converter.representative_dataset = representative_dataset

model_tflite = converter.convert()

# Save the model to disk
open(MODEL_TFLITE, "wb").write(model_tflite)
```
